```python
# Размер надписей на графиках
PLOT_LABEL_FONT_SIZE = 14
# Генерация цветовой схемы
# Возвращает список цветов
def getColors(n):
    COLORS = []
    cm = plt.cm.get_cmap('hsv', n)
    for i in np.arange(n):
        COLORS.append(cm(i))
    return COLORS
```

```python
df = combined.copy()
df['G1'] = df['G1'].apply(lambda x: x/1000000000)
df['G2'] = df['G2'].apply(lambda x: x/1000000000)
df['G3'] = df['G3'].apply(lambda x: x/1000000000)
df['doxod'] = df['doxod'].apply(lambda x: x/1000000000)

for i in range(len(df['year'])):
    if df.iloc[i, 6] > 2006 and df.iloc[i, 6] < 2011 and df.iloc[i, 5] > 0:
        df.iloc[i, 5] = df.iloc[i, 5]/100
    if df.iloc[i, 6] < 2006 and df.iloc[i, 6] < 2011 and df.iloc[i, 4] > 0:
        df.iloc[i, 4] = df.iloc[i, 4]/100
    if df.iloc[i, 6] < 2006 and df.iloc[i, 6] < 2011 and df.iloc[i, 3] > 0:
        df.iloc[i, 3] = df.iloc[i, 3]/100
    if df.iloc[i, 6] < 2006 and df.iloc[i, 6] < 2011 and df.iloc[i, 2] > 0:
        df.iloc[i, 2] = df.iloc[i, 2]/100
```
